```python
import folium

# Coordinates for Memphis, TN
latitude = 35.11
longitude = -89.97

# Create a map centered on Memphis, TN
m = folium.Map(location=[latitude, longitude], zoom_start=5)

# Convert the radius to meters (since folium uses meters)
radius_miles = 650
radius_meters = radius_miles * 1609.34  # 600 miles in meters

# Add a circle representing the 600-mile radius
folium.Circle(
    location=[latitude, longitude],
    radius=radius_meters,  # Radius in meters
    color='blue',
    fill=True,
    fill_opacity=0.3,
    fill_color='lightblue',
).add_to(m)

# Add a marker for Memphis
folium.Marker(
    location=[latitude, longitude],
    popup='Memphis, TN',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)

# Show the map
m
```
